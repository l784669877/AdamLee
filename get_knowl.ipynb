```python
from tqdm import tqdm
from PyPDF2 import PdfReader

def read_pdf_chunks(pdf_path, pages_per_chunk=5, overlap=3):
    pdf_reader = PdfReader(open(pdf_path, 'rb'))
    total_pages = len(pdf_reader.pages)
    chunks = []

    start_page = 0
    while start_page < total_pages:
        end_page = min(start_page + pages_per_chunk, total_pages)
        chunk = ""
        for page_num in range(start_page, end_page):
            page = pdf_reader.pages[page_num]
            chunk += page.extract_text()
        chunks.append(chunk)
        start_page += pages_per_chunk - overlap

    return chunks

pdf_path = '内置指令.pdf'
chunks = read_pdf_chunks(pdf_path, pages_per_chunk=5, overlap=1)

with open('temp.txt', 'a', encoding='utf-8') as file:
    for i, chunk in tqdm(enumerate(chunks), total=len(chunks)):
        temp = get_knowledge(chunk, glm4.bind(temperature=0.5))

        if temp != '无':
            file.write(temp)
            file.write("\n\n")
            file.flush()
```
